# Cancer Data Analysis

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../data/Cancer_Data.csv")
df.head()
```

```python
sns.set_theme()

sns.catplot(
    data=df,
    x="diagnosis", y="radius_mean", hue="diagnosis"
)
```

```python
m = df[df["diagnosis"] == "M"]
b = df[df["diagnosis"] == "B"]

data = {
    "diagnosis": ["B", "M"],
    "radius_mean": [b["radius_mean"].mean(), m["radius_mean"].mean()],
    "texture_mean": [b["texture_mean"].mean(), m["texture_mean"].mean()],
    "perimeter_mean": [b["perimeter_mean"].mean(), m["perimeter_mean"].mean()],
    "area_mean": [b["area_mean"].mean(), m["area_mean"].mean()],
    "smoothness_mean": [b["smoothness_mean"].mean(), m["smoothness_mean"].mean()],
    "compactness_mean": [b["compactness_mean"].mean(), m["compactness_mean"].mean()],
}

dfm = pd.DataFrame(data=data)
dfm = dfm.set_index("diagnosis")
dfm
```
